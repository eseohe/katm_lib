## Load 20 Newsgroups Subset

We load a subset of the 20 newsgroups dataset for a quick demonstration.

```python
from sklearn.datasets import fetch_20newsgroups

# Load a small subset for quick demonstration
newsgroups = fetch_20newsgroups(
    remove=('headers', 'footers', 'quotes'),
    subset='train'
)

documents = newsgroups.data[:500]  # Use first 500 documents for speed
print(f"Loaded {len(documents)} documents")
```

## Fit KATM with Default KeyBERT Extractor

KATM uses KeyBERT by default for keyphrase extraction.

```python
# Initialize and fit KATM with default KeyBERT extractor
model = KATM(
    kp_algorithm='keybert',
    n_topics=10,
    n_keyphrases=10
)

model.fit(documents)
print("KATM model fitted successfully!")
```

## Print Topics

```python
model.print_topics(n_words=10)
```

## Algorithm Comparison: RAKE and YAKE

KATM supports multiple keyphrase extraction algorithms. Let's compare RAKE and YAKE.

```python
# Fit KATM with RAKE
model_rake = KATM(
    kp_algorithm='rake',
    n_topics=10,
    n_keyphrases=10
)
model_rake.fit(documents)
print("KATM with RAKE fitted!")
model_rake.print_topics(n_words=10)
```

```python
# Fit KATM with YAKE
model_yake = KATM(
    kp_algorithm='yake',
    n_topics=10,
    n_keyphrases=10
)
model_yake.fit(documents)
print("KATM with YAKE fitted!")
model_yake.print_topics(n_words=10)
```

## Transform New Documents

The `transform()` method allows you to get topic distributions for new documents without re-fitting.

```python
# Create a held-out document
held_out_doc = [
    "Space exploration has led to many scientific discoveries about our solar system and beyond."
]

# Get topic distribution for the held-out document
topic_probs = model.transform(held_out_doc)
print(f"Topic distribution shape: {topic_probs.shape}")
print(f"Topic probabilities: {topic_probs[0]}")
```

## Fit-Transform Combined

You can also use `fit_transform()` to fit and get topic distributions in one step.

```python
# Get topic distributions for training documents
doc_topic_matrix = model.fit_transform(documents[:10])
print(f"Document-topic matrix shape: {doc_topic_matrix.shape}")
```
